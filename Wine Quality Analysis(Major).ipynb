#### Plotting histograms for numeric features

```python
numeric_features = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
                    'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
                    'pH', 'sulphates', 'alcohol', 'quality']

for feature in numeric_features:
    plt.figure(figsize=(5, 3))
    sns.histplot(data=df, x=feature, kde=True)
    plt.title(f'Distribution of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.show()
```

#### Relationships between features and the target variable ('quality') using scatter plots & other visualizations.


```python
# Scatter plot of each feature against 'quality'
for feature in numeric_features[:-1]:  # Exclude 'quality'
    plt.figure(figsize=(5, 3))
    sns.scatterplot(data=df, x=feature, y='quality')
    plt.title(f'{feature} vs quality')
    plt.xlabel(feature)
    plt.ylabel('quality')
    plt.show()
```

Make just 2 categories good and bad.

```python
ranges = (2,6.5,8) 
groups = ['bad','good']
df['quality'] = pd.cut(df['quality'],bins=ranges,labels=groups)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['quality'] = le.fit_transform(df['quality'])
df.head()
```

```python
df['quality'].value_counts()
```

#### Balancing Imbalanced Dataset

```python
good_quality = df[df['quality']==1]
bad_quality = df[df['quality']==0]

bad_quality = bad_quality.sample(frac=1)
bad_quality = bad_quality[:217]

new_df = pd.concat([good_quality,bad_quality])
new_df = new_df.sample(frac=1)
new_df
```

```python
new_df['quality'].value_counts()
```

#### Checking the correlation between columns

```python
new_df.corr()['quality'].sort_values(ascending=False)
```

### Splitting the data into train and test

```python
from sklearn.model_selection import train_test_split
X = new_df.drop('quality',axis=1) 
y = new_df['quality']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
param = {'n_estimators':[100,200,300,400,500,600,700,800,900,1000]}

grid_rf = GridSearchCV(RandomForestClassifier(),param,scoring='accuracy',cv=10,)
grid_rf.fit(X_train, y_train)
```

```python
pred = grid_rf.predict(X_test)
```

```python
print(confusion_matrix(y_test,pred))
```

```python
print(classification_report(y_test,pred))
```

```python
print(accuracy_score(y_test,pred))
```
